# Projeto de Análise de Dados – Casos de Dengue no Brasil

## Objetivo

Este projeto tem como objetivo analisar a evolução dos casos de dengue no Brasil utilizando técnicas de Análise Exploratória de Dados (EDA). A partir de dados públicos, busca-se identificar padrões temporais e geográficos, contribuindo para a compreensão da distribuição da doença e demonstrando como a análise de dados pode apoiar a tomada de decisões na área da saúde pública.

## Problema

A dengue representa um importante problema de saúde pública no Brasil, causando milhares de casos todos os anos e impactando diretamente a população e o sistema de saúde. A análise de dados permite identificar estados mais afetados, períodos de maior incidência e tendências ao longo do tempo, fornecendo informações relevantes para ações de prevenção e controle da doença.

Importar bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

Ler o arquivo

```python
df = pd.read_csv("dengue_brazil.csv")

df.head()
```

Conhecendo os dados

```python
df.info()
```

Estatísticas

```python
df.describe()
```

Valores nulos

```python
df.isnull().sum()
```

Transformar os dados

```python
df_long = df.melt(
    id_vars="time",
    var_name="estado",
    value_name="casos"
)

df_long.head()
```

## Tratamento dos Dados

Nesta etapa foi realizada a transformação da base de dados do formato largo (uma coluna para cada estado) para o formato longo, utilizando a função `melt()` do Pandas.

Também foi realizada a conversão da coluna de datas para o tipo `datetime`, permitindo análises temporais e a criação da variável **ano**, utilizada nas análises e visualizações.

Converter a data

```python
df_long["time"] = pd.to_datetime(df_long["time"])

df_long.head()
```

Criar coluna Ano

```python
df_long["ano"] = df_long["time"].dt.year

df_long.head()
```

Conferir

```python
df_long.info()
```

Casos por estado

```python
estado_total = (
    df_long
    .groupby("estado")["casos"]
    .sum()
    .sort_values(ascending=False)
)

estado_total
```

Top 10

```python
estado_total.head(10)
```

Primeiro gráfico

```python
plt.figure(figsize=(12,6))

estado_total.head(10).plot(kind="bar")

plt.title("Top 10 Estados com Maior Número de Casos de Dengue")

plt.xlabel("Estado")

plt.ylabel("Casos")

plt.show()
```

Casos por ano

```python
casos_ano = (
    df_long
    .groupby("ano")["casos"]
    .sum()
)

casos_ano
```

Gráfico temporal

```python
plt.figure(figsize=(12,6))

casos_ano.plot()

plt.title("Evolução dos Casos de Dengue")

plt.xlabel("Ano")

plt.ylabel("Casos")

plt.show()
```

Média por estado

```python
media_estado = (
    df_long
    .groupby("estado")["casos"]
    .mean()
    .sort_values(ascending=False)
)

media_estado.head(10)
```

Percentual dos 5 estados com mais casos

```python
top5 = estado_total.head(5)

percentual = (top5 / estado_total.sum()) * 100

percentual
```

Ano com mais casos

```python
casos_ano.sort_values(ascending=False)
```

Percentual dos 3 estados com mais casos

```python
total_casos = estado_total.sum()

top3 = (estado_total.head(3) / total_casos * 100).round(2)

print("Percentual dos 3 estados com mais casos:")
print(top3)
```

Diferença entre o maior e o menor estado

```python
print(f"Estado com mais casos: {estado_total.idxmax()} - {estado_total.max():,.0f} casos")
print(f"Estado com menos casos: {estado_total.idxmin()} - {estado_total.min():,.0f} casos")
```

Salvar o arquivo tratado

```python
df_long.to_csv("dengue_tratado.csv", index=False)

print("Arquivo salvo com sucesso!")
```

# Conclusões da Análise

A análise dos dados permitiu identificar diferenças significativas na incidência de dengue entre os estados brasileiros no período de **2000 a 2019**.

O estado de **São Paulo** apresentou o maior número de casos registrados, com **2.535.416 casos**, seguido por **Minas Gerais**, com **2.285.503 casos**, e **Rio de Janeiro**, com **1.412.425 casos**. Esses resultados demonstram uma forte concentração da doença nas regiões Sudeste e Centro-Oeste.

Em contrapartida, os menores volumes de casos foram observados em **Rio Grande do Sul (12.851 casos)**, **Santa Catarina (13.760 casos)**, **Roraima (39.127 casos)** e **Amapá (40.176 casos)**, evidenciando diferenças importantes na distribuição geográfica da dengue no Brasil.

A análise temporal mostrou que **2015** foi o ano com maior número de registros, totalizando **1.700.248 casos**, seguido por **2019 (1.558.424 casos)** e **2016 (1.514.749 casos)**. Esses picos indicam períodos de maior disseminação da doença, possivelmente influenciados por fatores climáticos, ambientais e pelo aumento da população do mosquito transmissor.

Observou-se também uma grande variação na quantidade de casos ao longo dos anos, reforçando o comportamento sazonal da dengue e a necessidade de monitoramento contínuo por parte dos órgãos de saúde.

Os resultados obtidos demonstram que a análise de dados é uma ferramenta essencial para identificar regiões prioritárias, acompanhar a evolução da doença ao longo do tempo e apoiar a tomada de decisões para ações de prevenção, controle epidemiológico e melhor distribuição de recursos públicos.
