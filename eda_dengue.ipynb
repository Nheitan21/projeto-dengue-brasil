```python
import matplotlib.pyplot as plt

df.iloc[:, 0:2].plot()
plt.title("Evolução inicial da dengue")
plt.show()
```

```python
df_long = df.melt(id_vars=['time'], var_name='estado', value_name='casos')
df_long.head()
```

```python
df_long['time'] = pd.to_datetime(df_long['time'], errors='coerce')
df_long.head()
```

```python
df_long.info()
```

```python
df_long.groupby('estado')['casos'].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

df_long.groupby('estado')['casos'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top 10 Estados com mais casos de Dengue')
plt.ylabel('Casos')
plt.xlabel('Estado')
plt.show()
```

```python
df_long.groupby('time')['casos'].sum().plot(figsize=(12,5))
plt.title('Evolução dos casos de Dengue no Brasil')
plt.ylabel('Casos')
plt.xlabel('Tempo')
plt.show()
```

```python
df_long.groupby('estado')['casos'].mean().sort_values(ascending=False).head(10)
```

## 📌 Insights principais

- Os casos de dengue variam significativamente entre os estados brasileiros  
- Estados como São Paulo, Minas Gerais e Bahia concentram maior volume de casos  
- Existe uma tendência de variação ao longo do tempo, com períodos de aumento e queda  
- Isso indica comportamento sazonal da doença  
- A distribuição desigual reforça a necessidade de políticas regionais de prevenção
